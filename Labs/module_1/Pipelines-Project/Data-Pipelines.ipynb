```python
bag_words = [j for i in df.Activity.str.lower().dropna().values for j in i.split()]
```

```python
bag_words = [j for i in df.Activity.str.lower().dropna().values for j in re.findall(r'[a-z_]{3,}',i)]
```

```python
print(len(bag_words))
print(len(set(bag_words)))
```

```python
words=list(set(bag_words))
words
```

```python
words=list(map(lambda x: re.fullmatch(r'(.+?)(?:ing)?(?:ed)?(?:er)?\b',x).group(1),bag_words))
```

```python
len(words)
```

```python
pd.Series(words).value_counts().head(30)
```

```python
categories = set(pd.Series(words).value_counts().head(30).index)
```

```python
categories=categories - set(['aft','for','free','from','his','into','sea','shark','swimm','the','was','wat','with']) | set(['swim'])
```

```python
len(categories)
```

```python
categories
```

```python
df=df.Activity.fillna('missing value')
```

```python
for i in categories:
    df.loc[df.Activity.str.lower().str.contains(i), 'New_Activity']=i
```
